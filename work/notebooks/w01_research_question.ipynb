## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import os, sys, subprocess
import pandas as pd

# 1. Colab Setup (downloads the data if it's not there)
IN_COLAB = "google.colab" in sys.modules
REPO_DIR = "flyrank-ml-internship-starter"

if IN_COLAB:
    if not os.path.isdir(REPO_DIR):
        subprocess.run(["git", "clone", "--depth", "1", "https://github.com/flyrank-bih/flyrank-ml-internship-starter", REPO_DIR], check=True)
    if os.path.basename(os.getcwd()) != REPO_DIR:
        os.chdir(REPO_DIR)
elif os.path.basename(os.getcwd()) == "notebooks":
    os.chdir("..") # move up to root if running locally

# 2. Now we can safely load the data!
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# 3. Calculate our numbers
declining_count = (df['trend_direction'] == 'down').sum()
total_impressions_down = df[df['trend_direction'] == 'down']['impressions_90d'].sum()

print(f"1. There are {declining_count:,} content items currently showing a downward trend.")
print(f"2. These declining pages received {total_impressions_down:,.0f} impressions in the last 90 days, which are at risk.")
print(f"3. If an editor can only refresh 50 pages a week, a ranking model ensures they target the highest value ones.")
```
